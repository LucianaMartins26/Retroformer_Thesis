```python
import pandas as pd

csv_files = [
    '100_epochs_accuracies.csv',
    '300_epochs_accuracies.csv',
    '500_epochs_accuracies.csv',
    '700_epochs_accuracies.csv',
    '700_epochs_accuracies.csv'
]

dataframes = [pd.read_csv(file).drop(columns=['Unnamed: 0'], errors='ignore') for file in csv_files]

combined_accuracies_dataframe = pd.concat(dataframes, ignore_index=True)

combined_accuracies_dataframe = combined_accuracies_dataframe.round(3)
combined_accuracies_dataframe
```

```python
import pandas as pd

test_files = [
    'output_100_epochs_test_data.txt', 'output_300_epochs_test_data.txt',
    'output_500_epochs_test_data.txt', 'output_700_epochs_test_data.txt', 
    'output_1000_epochs_test_data.txt'
]
train_files = [
    'output_100_epochs_train_data.txt', 'output_300_epochs_train_data.txt',
    'output_500_epochs_train_data.txt', 'output_700_epochs_train_data.txt',
    'output_1000_epochs_train_data.txt'
]
val_files = [
    'output_100_epochs_val_data.txt', 'output_300_epochs_val_data.txt',
    'output_500_epochs_val_data.txt', 'output_700_epochs_val_data.txt',
    'output_1000_epochs_val_data.txt'
]

data_train = {'Model': [], 'Top-1 Train': [], 'Top-3 Train': [], 'Top-5 Train': [], 'Top-10 Train': []}
data_val = {'Model': [], 'Top-1 Val': [], 'Top-3 Val': [], 'Top-5 Val': [], 'Top-10 Val': []}
data_test = {'Model': [], 'Top-1 Test': [], 'Top-3 Test': [], 'Top-5 Test': [], 'Top-10 Test': []}

def extract_top_values(file_path):
    top_values = {'Top-1': None, 'Top-3': None, 'Top-5': None, 'Top-10': None}
    with open(file_path, 'r') as file:
        for line in file:
            for key in top_values.keys():
                if key in line:
                    parts = line.split(':')
                    if len(parts) == 2:
                        metric = parts[0].strip()
                        value = parts[1].strip()
            
                        try:
                            value = float(value)
                        except ValueError:
                            value = None
                        
                        if metric in top_values:
                            top_values[metric] = value
    return top_values

for train_file, val_file, test_file in zip(train_files, val_files, test_files):
    epochs = train_file.split('_')[1]
    
    train_values = extract_top_values(train_file)
    val_values = extract_top_values(val_file)
    test_values = extract_top_values(test_file)
    
    data_train['Model'].append(f"{epochs} epochs")
    data_val['Model'].append(f"{epochs} epochs")
    data_test['Model'].append(f"{epochs} epochs")
    
    for key in ['Top-1', 'Top-3', 'Top-5', 'Top-10']:
        data_train[f'{key} Train'].append(round(train_values[key],3))
        data_val[f'{key} Val'].append(round(val_values[key],3))
        data_test[f'{key} Test'].append(round(test_values[key],3))

df_train = pd.DataFrame(data_train)
df_val = pd.DataFrame(data_val)
df_test = pd.DataFrame(data_test)
```
